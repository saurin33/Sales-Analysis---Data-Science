```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

# Merging 12 months of data

```python
df =  pd.read_csv('Sales_Data/Sales_January_2019.csv')

files = [file for file in os.listdir('Sales_Data')]
for file in files:
    print(file)
```

```python
all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv('Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data
```

```python
all_months_data.to_csv('all_data.csv', index=False)
```

```python
all_months_data.isnull().sum()
```

```python
all_months_data['Month'] = all_months_data['Order Date'].str[0:2]
all_months_data.head()
```

```python
nan_df = all_months_data[all_months_data.isna().any(axis=1)]
nan_df
```

```python
all_data = all_months_data.dropna(how='all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
all_data['Month'] = all_data['Month'].astype(int)
all_data
```

```python
all_data.info()
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

What Was the best month for sell? 

```python
results = all_data.groupby('Month').sum()
results
```

```python
months = range(1, 13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Months')
plt.show()
```

# What City had highest sales? 

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

```python
city_results = all_data.groupby('City').sum()
city_results
```

```python
cities = [city for city, df in all_data.groupby('City')]
cities
```

```python
plt.bar(cities, city_results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.xlabel('Citi name')
plt.ylabel('Sales in USD')
plt.show()
```

```python
city_results_df = pd.DataFrame(city_results)
city_results_df
```

```python
city_clean_df = city_results_df.drop(columns=['Quantity Ordered', 'Price Each', 'Month'])
city_clean_df
```

```python
plt.bar(city_results.index, city_results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.xlabel('Citi name')
plt.ylabel('Sales in USD')
plt.show()
```

## What time do the businesses need to run the advertisement? 

```python
# all_data = all_data.drop(columns='Date')
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()
```

```python
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hourly_results = all_data.groupby('Hour').sum()
hourly_results_df = pd.DataFrame(hourly_results)
hourly_results_df.head()
```

```python
plt.plot(hourly_results_df.index, hourly_results_df['Sales'])
plt.xticks(hourly_results_df.index)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
plt.title('Hourly Sales in USD ')
plt.grid()
plt.show()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
hours
```

```python
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.show()
```

# What products are sold together

```python
all_data.head()
```

```python
products_df = all_data.groupby('Order ID')
products_df
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key, value)
```

```python
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,3)))

for key, value in count.most_common(10):
    print(key, value)
```

# What Products Sold the most? 

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered_df = pd.DataFrame(quantity_ordered)
quantity_ordered_df
```

```python
products = [product for product, df in product_group]
products
```

```python
plt.bar(quantity_ordered_df.index, quantity_ordered_df['Quantity Ordered'])
plt.xticks(quantity_ordered_df.index, rotation='vertical', size = 8)
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
prices
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(quantity_ordered_df.index, quantity_ordered_df['Quantity Ordered'])
ax2.plot(quantity_ordered_df.index, prices, 'g-')

ax1.set_xticklabels(quantity_ordered_df.index, rotation='vertical', size = 8)

ax1.set_xlabel('Products Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price', color='b')

plt.show()
```
